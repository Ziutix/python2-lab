Ćwiczenie 1: Series — tworzenie i operacje

```python
import pandas as pd
import numpy as np

# Series z listy — domyślny indeks (0, 1, 2, ...)
oceny = pd.Series([4.5, 3.0, 5.0, 3.5, 4.0])
print(oceny)
```

```python
#Zadanie 1a: Utwórz Series pensje z dict:
pensje = pd.Series({
    'Anna': 5500,
    'Jan': 7200,
    'Ewa': 4800,
    'Marek': 9100,
    'Kasia': 6300
})

print(pensje)
```

```python
#Zadanie 1b: Odpowiedz na pytania o pensje:
# 1. kto zarabia najwięcej
print("Najwięcej zarabia:", pensje.idxmax())

# 2. średnia pensja
print("Średnia pensja:", pensje.mean())

# 3. osoby powyżej średniej
print("Powyżej średniej:")
print(pensje[pensje > pensje.mean()])

# 4. ile osób zarabia poniżej 6000
print("Liczba osób < 6000:", (pensje < 6000).sum())
```

```python
#Zadanie 1c: Oblicz pensje po zmianach:
# 1. podwyżka 10%
pensje_10 = pensje * 1.10
print("Po podwyżce 10%:\n", pensje_10)

# 2. premia 500 zł
pensje_premia = pensje + 500
print("\nPo premii 500:\n", pensje_premia)

# 3. nowa pensja (podwyżka + premia)
pensje_nowe = pensje * 1.10 + 500
print("\nNowe pensje:\n", pensje_nowe)

# 4. ranking od najwyższej
ranking = pensje_nowe.sort_values(ascending=False)
print("\nRanking (od najwyższej):\n", ranking)
```

```python
#Zadanie 2b: Dodaj nowe kolumny do DataFrame sklep:
# 1. wartość magazynu
sklep['wartosc_magazynu'] = sklep['cena'] * sklep['stan_magazynowy']

# 2. cena z VAT
sklep['cena_z_vat'] = sklep['cena'] * 1.23

# 3. cały DataFrame
print(sklep)
```

Ćwiczenie 3: Eksploracja datasetu — samodzielna analiza

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(url)
```

```python
#Zadanie 3a: Pierwszy rzut oka

# 1. head(10)
print(penguins.head(10))

# 2. shape
print("\nShape:", penguins.shape)

# 3. info()
print("\nInfo:")
penguins.info()
```

```python
#Zadanie 3b: Statystyki opisowe

# 1. describe()
print(penguins.describe())

# 2. describe(all)
print("\nDescribe all:\n", penguins.describe(include='all'))

# 3. mediana masy
print("\nMediana body_mass_g:", penguins['body_mass_g'].median())
```

```python
#Zadanie 3c: Rozkłady wartości
# 1. ile gatunków
print("Species:\n", penguins['species'].value_counts())

# 2. ile na wyspach
print("\nIsland:\n", penguins['island'].value_counts())

# 3. płeć
print("\nSex:\n", penguins['sex'].value_counts())

# 4. % Adelie
print("\n% Adelie:\n", penguins['species'].value_counts(normalize=True))
```

```python
#Zadanie 3d: Brakujące dane
# 1. brakujące wartości
print("Braki:\n", penguins.isna().sum())

# 2. % braków w sex
braki_sex = penguins['sex'].isna().sum()
procent = braki_sex / len(penguins) * 100

print(f"\n% braków w 'sex': {procent:.2f}%")
```

Wnioski:
-Dataset zawiera 344 pingwiny z 3 gatunków (Adelie, Chinstrap, Gentoo).
-Najczęstszy gatunek to Adelie (152 osobniki, ok. 44%).
-Pingwiny występują na 3 wyspach, najwięcej na Biscoe (168 osobników).
-W danych występują braki – np. w kolumnie sex brakuje ok. 3% wartości.
-Średnia masa ciała pingwinów wynosi ok. 4200 g, a mediana to ok. 4050 g.
-Dane dotyczące płci nie są idealnie zbalansowane (więcej samców niż samic).

Ćwiczenie 4: Pytania biznesowe + commit

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

# Dodaj kolumnę z procentem napiwku
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)
```

```python
#Zadanie 4a:
# 1. średni rachunek
print("Średni rachunek:", tips['total_bill'].mean())

# 2. średni napiwek %
print("Średni napiwek %:", tips['tip_pct'].mean())

# 3. który dzień ma najwięcej rachunków
print("\nRachunki wg dnia:\n", tips['day'].value_counts())

# 4. ile rachunków od palaczy
print("\nSmoker:\n", tips['smoker'].value_counts())

# 5. describe dla tip + mediana
print("\nDescribe tip:\n", tips['tip'].describe())
print("Mediana napiwku:", tips['tip'].median())

# 6. % Dinner vs Lunch
print("\nProcent Lunch/Dinner:\n", tips['time'].value_counts(normalize=True) * 100)

# 7. ile unikalnych rozmiarów grup
print("\nUnikalne size:", tips['size'].nunique())
```
