Ćwiczenie 3: Operacje wektorowe — analiza biznesowa 

```python
# Produkty w sklepie internetowym
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
         'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']

ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

```python
#Zadanie 1: Podstawowe statystyki cenowe
# Średnia cena
print("Średnia cena:", ceny.mean())

# Najdroższy produkt
idx_max = ceny.argmax()
print("Najdroższy:", nazwy[idx_max], ceny[idx_max])

# Najtańszy produkt
idx_min = ceny.argmin()
print("Najtańszy:", nazwy[idx_min], ceny[idx_min])

```

```python
#Zadanie 2: Wartość magazynu
# Wartość każdego produktu
wartosci = ceny * ilosci_magazyn

print("Wartość produktów:", wartosci)

# Łączna wartość magazynu
print("Łączna wartość magazynu:", wartosci.sum())
```

```python
#Zadanie 3: Analiza cenowa

# Rabat 15%
ceny_rabat = ceny * 0.85
print("Ceny po rabacie:", ceny_rabat)

# VAT 23%
ceny_vat = ceny * 1.23
print("Ceny z VAT:", ceny_vat)

# Ile produktów > 1000 zł
ile_drogich = (ceny > 1000).sum()
print("Produkty powyżej 1000 zł:", ile_drogich)
```

```python
#Zadanie 4: Filtrowanie
nazwy_np = np.array(nazwy)
# Produkty z oceną >= 4.5
maska = oceny_klientow >= 4.5
print("Produkty (>=4.5):", nazwy_np[maska])
print("Ceny:", ceny[maska])

# Produkty poniżej 200 zł
maska = ceny < 200
print("Produkty (<200 zł):", nazwy_np[maska])
print("Ilości:", ilosci_magazyn[maska])
```

```python
#Zadanie 5: Analiza zapasów
# Produkty z niskim stanem (< 30)
maska = ilosci_magazyn < 30
print("Niski stan:", nazwy_np[maska])

# Wartość tych produktów
wartosci = ceny * ilosci_magazyn
print("Wartość niskiego stanu:", wartosci[maska].sum())

# Procent całego magazynu
procent = wartosci[maska].sum() / wartosci.sum() * 100
print("Procent magazynu:", procent)
```

```python
#Zadanie 6 (bonus): Analiza korelacji cena–ocena
# Okazje (ocena >= 4.5 i cena < 1000)
maska = (oceny_klientow >= 4.5) & (ceny < 1000)
print("Okazje:", nazwy_np[maska])
print("Ceny:", ceny[maska])

# Średnia cena dla ocen 4.0–4.5
maska = (oceny_klientow >= 4.0) & (oceny_klientow < 4.5)
print("Średnia cena:", ceny[maska].mean())
```

Ćwiczenie 4: Benchmark — lista vs NumPy

```python
import time

# Przygotowanie danych
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)

# Test 1: Pomnóż przez 2
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start

print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 2: Suma
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start

start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start

print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 3: Filtrowanie (elementy > 500000)
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start

print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```
