Ćwiczenie 2: Reshape i łączenie danych

```python
#Zadanie 2a — Reshape

dane_flat = np.array([120, 135, 98, 142, 167, 189, 201, 178, 156, 143, 198, 221])

kwartaly = dane_flat.reshape(4, 3)

print("Kwartały:\n", kwartaly)

suma_kw = kwartaly.sum(axis=1)
print("Suma per kwartał:", suma_kw)

najlepszy = suma_kw.argmax() + 1
print("Najlepszy kwartał: Q", najlepszy)
```

```python
#Zadanie 2b — Stacking
warszawa = np.array([150, 160, 170, 180])
krakow = np.array([120, 130, 140, 150])
wroclaw = np.array([90, 100, 110, 120])

wszystkie = np.vstack([warszawa, krakow, wroclaw])

print("Wszystkie:\n", wszystkie)

print("Suma per oddział:", wszystkie.sum(axis=1))
print("Suma per kwartał:", wszystkie.sum(axis=0))
```

```python
#Zadanie 2c — Praktyczne
sprzedaz_flat = np.array([100, 200, 150, 300, 250, 180, 120, 220, 170, 310, 260, 190])

# 1. reshape 6x2
macierz = sprzedaz_flat.reshape(6, 2)

print("Macierz:\n", macierz)

# 2. średnia sprzedaż per produkt
srednia = macierz.mean(axis=1)
print("Średnia per produkt:", srednia)

# 3. zmiana między miesiącami
zmiana = macierz[:, 1] - macierz[:, 0]
print("Zmiana miesiąc 2 - 1:", zmiana)
```

```python
#Zadanie 2d — Flatten i concatenate
# zakładam że masz już kwartaly z zadania 2a

wektor = kwartaly.flatten()
print("Flatten:", wektor)

nowy = np.concatenate([wektor, np.array([235])])
print("Po dodaniu miesiąca:", nowy)
 
```

Ćwiczenie 3: Analiza danych finansowych

```python
np.random.seed(2026)

# 8 pracowników, dane roczne
pracownicy = ['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek']
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.0, 4.7, 3.5, 4.3, 3.9])

# Sprzedaż miesięczna 4 produktów przez 6 miesięcy
produkty_nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Akcesorium']
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_produktow = np.array([3500, 1800, 2500, 150])
```

```python
#Zadanie 3a: Analiza wynagrodzeń
# 1. statystyki
srednia = pensje.mean()
mediana = np.median(pensje)
odchylenie = pensje.std()

print("Średnia:", srednia)
print("Mediana:", mediana)
print("Odchylenie std:", odchylenie)


# 2. osoby powyżej średniej
pracownicy_np = np.array(pracownicy)
powyzej = pracownicy_np[pensje > srednia]

print("Powyżej średniej:", powyzej)


# 3. osoby poniżej mediany
ponizej = pracownicy_np[pensje < mediana]

print("Poniżej mediany:", ponizej)


# 4. rozstęp i IQR
rozstep = pensje.max() - pensje.min()
iqr = np.percentile(pensje, 75) - np.percentile(pensje, 25)

print("Rozstęp:", rozstep)
print("IQR:", iqr)
```

```python
#Zadanie 3b: System premii (np.where)
# premie
premie = np.where(oceny >= 4.5, 2000,
         np.where(oceny >= 4.0, 1000, 0))

print("Premie:", premie)

# nowe wynagrodzenie
nowe_pensje = pensje + premie
print("Nowe pensje:", nowe_pensje)

# kto dostał największą premię
idx = premie.argmax()
print("Najwyższa premia:", pracownicy[idx])
```

```python
#Zadanie 3c: Ranking (argsort);
kolejnosc = np.argsort(pensje)[::-1]

for i, idx in enumerate(kolejnosc, 1):
    print(i, pracownicy[idx], pensje[idx], staz[idx])
```

```python
#Zadanie 3d: Korelacja
# korelacje
korelacja_staz_pensja = np.corrcoef(staz, pensje)[0, 1]
korelacja_staz_ocena = np.corrcoef(staz, oceny)[0, 1]

print("Staż vs pensja:", korelacja_staz_pensja)
print("Staż vs ocena:", korelacja_staz_ocena)

# który silniejszy
if abs(korelacja_staz_pensja) > abs(korelacja_staz_ocena):
    print("Silniejsza zależność: staż vs pensja")
else:
    print("Silniejsza zależność: staż vs ocena")
```

```python
#Zadanie 3e: Analiza sprzedaży (broadcasting)
# 1. poprawne dopasowanie wymiarów
ceny_kol = ceny_produktow.reshape(4, 1)

# 2. przychód (4 produkty × 6 miesięcy)
przychod = sprzedaz * ceny_kol

print("Przychód:\n", przychod)


# 3. który produkt zarobił najwięcej (sumujemy po miesiącach)
suma_produkt = przychod.sum(axis=1)
idx_prod = suma_produkt.argmax()
print("Najlepszy produkt:", produkty_nazwy[idx_prod])


# 4. który miesiąc był najlepszy (sumujemy po produktach)
suma_miesiac = przychod.sum(axis=0)
idx_mies = suma_miesiac.argmax()
print("Najlepszy miesiąc:", idx_mies + 1)


# 5. trend (pierwsze 3 vs ostatnie 3 miesiące)
pierwsze = przychod[:, :3].sum()
ostatnie = przychod[:, 3:].sum()

print("Pierwsze 3 mies:", pierwsze)
print("Ostatnie 3 mies:", ostatnie)

if ostatnie > pierwsze:
    print("Trend rosnący 📈")
else:
    print("Trend malejący 📉")
```

Ćwiczenie 4: Generowanie danych

```python
 #Krok 1 — Generowanie
import numpy as np

np.random.seed(42)

# wiek: normalny, średnia 35, std 10, min 18
wiek = np.clip(np.random.normal(35, 10, 100).astype(int), 18, None)

# wydatki: normalny, średnia 500, std 200, min 0
wydatki = np.clip(np.random.normal(500, 200, 100), 0, None)

# liczba wizyt: Poisson (lambda = 5)
wizyty = np.random.poisson(5, 100)

print("Wiek:", wiek[:5])
print("Wydatki:", wydatki[:5])
print("Wizyty:", wizyty[:5])
```

```python
#Krok 2 — Analiza
# 1. Statystyki opisowe każdej zmiennej (mean, median, std, min, max)
print(f"Wiek → średnia: {wiek.mean():.1f}, mediana: {np.median(wiek):.1f}, std: {wiek.std():.1f}, min: {wiek.min()}, max: {wiek.max()}")
print(f"Wydatki → średnia: {wydatki.mean():.1f}, mediana: {np.median(wydatki):.1f}, std: {wydatki.std():.1f}, min: {wydatki.min():.1f}, max: {wydatki.max():.1f}")
print(f"Wizyty → średnia: {wizyty.mean():.1f}, mediana: {np.median(wizyty):.1f}, std: {wizyty.std():.1f}, min: {wizyty.min()}, max: {wizyty.max()}")


# 2. Korelacja wiek vs wydatki
print("Korelacja wiek vs wydatki:", np.corrcoef(wiek, wydatki)[0, 1])


# 3. Korelacja wizyty vs wydatki
print("Korelacja wizyty vs wydatki:", np.corrcoef(wizyty, wydatki)[0, 1])


# 4. Segmentacja klientów za pomocą np.where
sr = wydatki.mean()
sd = wydatki.std()

segment = np.where(wydatki > sr + sd, 'WYSOKI',
          np.where(wydatki < sr - sd, 'NISKI', 'ŚREDNI'))

print("Segmenty:", segment)


# 5. Ile klientów w każdym segmencie (np.unique)
unikalne, liczby = np.unique(segment, return_counts=True)

for s, l in zip(unikalne, liczby):
    print(s, ":", l)
```
